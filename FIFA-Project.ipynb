### 5 - Choosing the columns for our study

### 6 - Transforming data types & applying fixes to the dataset

Converting columns 'Value', 'Wage' and 'Release Clause' to integer. A bit of cleaning required. 

```python
# Column Value to Integer and cleaning numbers and letter to standarize the values.

fifa['Value'] = fifa['Value'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
```

```python
# Column Wage to Integer and cleaning numbers and letter to standarize the values.

fifa['Wage'] = fifa['Wage'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
```

```python
# Column Release CLause to Integer and cleaning numbers and letter to standarize the values.

fifa['Release Clause'] = fifa['Release Clause'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
```

```python
# Columns to integer again to minimize errors.

fifa['Value']=fifa['Value'].astype(int)
fifa['Wage']=fifa['Wage'].astype(int)
fifa['Release Clause']=fifa['Release Clause'].astype(int)
```

# Final Analysis

## 1 - First question

Our client is a football club in Spain whose main CB cannot play due to an injury. As they are mid-season, they cannot risk a negotiation and they are willing to pay the Release Clause for a new player.
Total budget is 40M€. Their first option is a low risk operation, buying the best player for the budget they have.
But also there is another condition: the player must come from the EU, since they cannot use any more foreigners.

### Considerations for our study

1. We decided to filter the dataset and include only European countries.
2. We placed a maximum budget condition on the "Release Clause column".
3. Decided to choose players from 25 years old and below, since the distribution analysis shows that 25 and above is when soccer player careers start to grow slower in overall quality, and therefore, the most growth potential as been already adquired. Choosing a player with a very high overall quality on their 25 years old, warantees we will have a good player for some years, and we wont need to replace him on the short run.

```python
EU_list = ("Austria", "Belgium", "Bulgaria", "Croatia", "Cyprus", "Czech Republic", "Denmark", "Estonia", "Finland", "France", "Germany", "Greece", "Hungary", "Republic of Ireland", "Italy", "Latvia", "Lithuania", "Luxembourg", "Malta", "Netherlands", "Poland", "Portugal", "Romania", "Slovakia", "Slovenia", "Spain", "Sweden")
fifa.loc[(fifa['Release Clause']<40000000) & (fifa['Nationality'].isin(EU_list)) & (fifa['Age']<=25) & (fifa['BP']=='CB')].sort_values('OVA', ascending=False)[:10]
```

### Results

Top 10 players, 25 and below, whith the highest OVA below 40M budget. Low risk operation.

## 2 - Second question

During the presentation, they also considered the option of buying 1 or 2 younger players with an expected great future but also under the same conditions as before. Higher risk will take in place, since we already have a replacement in our team, wich is not as good and already old.


### Considerations for our study

1. We decided to create a new column which takes overall quality divided by age. This give us the % of quality by age that a player achieved during his career. The higher this number, the better growth potential he has.
2. We kept all coniserations we had before, but sorted the list by OVA/AGE. 

```python
fifa["OVA/Age"]=fifa['OVA']/fifa['Age']

fifa.loc[(fifa['Release Clause']<40000000) & (fifa['Nationality'].isin(EU_list)) & (fifa['Age']<=25) & (fifa['BP']=='CB')].sort_values('OVA/Age', ascending=False)[:10]
```

### Results

Top 10 players, 25 and below, whith the highest OVA/Age % below 40M budget. High chance to be cracks in the future, but not guaranteed. Higher risk operation.

## EXTRA

We decided to create a function for simulating how many OVA points will a player increase by media depending on actual Age and actual OVA.
Useful to understand what expected quality will the player will reach at certain point in the future.


```python
possible = fifa.groupby(["Age"]).agg({"OVA":"mean"})
def expected(age,ova):
    future_age = age + 5 
    expected_growth = possible.loc[future_age] - possible.loc[age]
    expected_ova = ova + expected_growth
    return expected_ova
age = int(input("age: "))
ova = int(input("ova: "))
expected(age,ova)
```
